```python
#  
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os, glob
import copy

#  
img_path = os.getenv("HOME") + "/aiffel/exploration/camera_sticker/images"
img_list = glob.glob(img_path + "/*.jpg") 

img_list = sorted(img_list)
    
# OpenCV   , resize
img_bgr_list = list()
for img in img_list:
    tmp = cv2.imread(img)
    w = int(tmp.shape[1])
    h = int(tmp.shape[0])
    tmp_resize = cv2.resize(tmp, (w - w%10, h - h%10))
    img_bgr_list.append(tmp_resize)


# OpenCV   BGR . RGB  .
img_rgb_list = list()
for bgr in img_bgr_list:
    img_rgb_list.append(cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB))

#   
img_show_list = list()
for rgb in img_rgb_list:
    img_show_list.append(copy.deepcopy(rgb))

#  
for show in img_show_list:
    plt.imshow(show)
    plt.xticks([])   # x  
    plt.yticks([])   # y  
    plt.show()
```

```python
#   face detection

#      HOG    .
# HOG            ,      .
#         .

import dlib   

# dlib  hog detector 
detector_hog = dlib.get_frontal_face_detector()   #-    detector 

# detector   bounding box .
dlib_rects_dict = dict()
img_recog_list = list()
img_ignor_list = list()
i = 0
for show in img_show_list:
    if not detector_hog(show, 1):
        img_ignor_list.append(i)
        i = i + 1
    else:
        dlib_rects_dict[i] = detector_hog(show, 1)
        img_recog_list.append(i)
        i = i + 1

#    
#    print(dlib_rects)
print("*  ")
for key, value in dlib_rects_dict.items():
    dlib_rects = value
    for dlib_rect in dlib_rects:
        l = dlib_rect.left()
        t = dlib_rect.top()
        r = dlib_rect.right()
        b = dlib_rect.bottom()
        
        cv2.rectangle(img_show_list[key], (l,t), (r,b), (0,255,0), 2, lineType=cv2.LINE_AA)
        
    #  
    plt.imshow(img_rgb_list[key])
    plt.xticks([])   # x  
    plt.yticks([])   # y  
    plt.show()
    
    #  
    plt.imshow(img_show_list[key])
    plt.xticks([])   # x  
    plt.yticks([])   # y  
    plt.show()

print("*   ")
for i in img_ignor_list:
    #  
    plt.imshow(img_rgb_list[i])
    plt.xticks([])   # x  
    plt.yticks([])   # y  
    plt.show()

    
```

```python
#   face landmark
# face landmark localization :    
# face landmark detection  bountding box    .

#  landmark  
model_path = os.getenv('HOME')+'/aiffel/exploration/camera_sticker/models/shape_predictor_68_face_landmarks.dat'
landmark_predictor = dlib.shape_predictor(model_path)

# landmark_predictor rgb dlib.rectangle  dlib.full_object_detection .
landmarks_dict = dict()
list_points_dict = dict()
for key, value in dlib_rects_dict.items():
    temp = dict()
    dlib_rects = value
    count = 0
    for dlib_rect in dlib_rects:
        if len(dlib_rects_dict[key]) == 1:
            points = landmark_predictor(img_rgb_list[key], dlib_rects_dict[key][0])
            landmarks_dict[key] = list(map(lambda p: (p.x, p.y), points.parts()))
        else:
            if count == len(dlib_rects_dict[key]) - 1:
                points = landmark_predictor(img_rgb_list[key], dlib_rects_dict[key][count])
                temp[count] = list(map(lambda p: (p.x, p.y), points.parts()))
                landmarks_dict[key] = temp
            else:
                points = landmark_predictor(img_rgb_list[key], dlib_rects_dict[key][count])
                temp[count] = list(map(lambda p: (p.x, p.y), points.parts()))
                count = count + 1

# points dlib.full_object_detection   parts()      . 
#    (x, y)      .  list_points tuple(x, y) 68  .
#      list_landmark 68    
#  
for key, value in landmarks_dict.items():
    if len(value) == 68:
        for i in range(len(value)):
            point = value[i]
            cv2.circle(img_show_list[key], point, 2, (0, 255, 255), -1) # yellow
    else:
        for i in list(value.keys()):
            v = value[i]
            for j in range(len(v)):
                point = v[j]
                cv2.circle(img_show_list[key], point, 2, (0, 255, 255), -1) # yellow


                 
#   
for i in img_recog_list:
    plt.imshow(img_show_list[i])
    plt.xticks([])   # x  
    plt.yticks([])   # y  
    plt.show()
```
